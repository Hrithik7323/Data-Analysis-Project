1. What was the Lead Actor/Actress of Title

```python
dataset.groupby('Lead Actor/Actress')['IMDB Rating'].count().plot(kind='bar', title=' Top Lead Actor/Actress of IMDB Rating')
# Show the graph
plt.xlabel("Number of Lead Actor/Actress")
plt.ylabel("Number of IMDB Rating")  # Fixed the ylabel to be more clear
plt.xticks(rotation=90)  # Added rotation for better readability
plt.tight_layout()  # Added to prevent label cutoff
plt.show()
```

2. Which was the best month for Movies IMDB Rating between 2023 to 2024?

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the dataset
dataset = pd.read_csv("hindi_movies_data.csv")

# Filtering data for 2023-2024
dataset_2023_2024 = dataset[(dataset['Year'] >= 2023) & (dataset['Year'] <= 2024)]

# Grouping data by month and IMDB Rating
dataset_2023_2024.groupby('Month')['IMDB Rating'].count().plot(kind='bar', color='skyblue', edgecolor='black')

# Enhanching the plot
plt.xlabel('Month', fontsize=12)
plt.ylabel('Number of Movies', fontsize=12)
plt.title('Best Month For Movies (2023-2024)', fontsize=14)
plt.xticks(rotation=0)   # Keeps month lables straight
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Adds a grid for better readability
plt.show()          
          
```

3 Which Director the higest Box Office Revenue (Crores) Year(2023-2024)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

dataset = pd.read_csv("hindi_movies_data.csv")

dataset_2023_2024 = dataset[(dataset['Year'] >= 2023) & (dataset['Year'] <= 2024)]

# Grouping data by Director and Box office 
top_Director = dataset_2023_2024.groupby('Director')['Box Office Revenue (Crores)'].sum().sort_values(ascending=False).head(10)

# Plot the top 10 Director
top_Director.plot(kind='bar', color='skyblue', edgecolor='black')

plt.xlabel('Director', fontsize=12)
plt.ylabel('Number of Box Office Revenue (Crores)', fontsize=12)
plt.title('Director Wise Top 10 Box Office Revenue (Crores) (2023-2024)', fontsize=14)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

4 What Movies Collection  the most in the two Year 2023 & 2024

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

dataset = pd.read_csv("hindi_movies_data.csv")

# Creating a Grid of Subplots (2*2 layout)
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# plt.subplots(2, 2) â†’ Creates a 2-row, 2-column grid for subplots.
# figsize=(12, 10) â†’ Defines the size of the entire figure.

# axs is an array of subplots where:
# axs[0,0] = Top left
# axs[0,1] = Top right
# axs[1,0] = Bottom left
# axs[1,1] = Bottom right (unused in this case)

top_MoviesCollection_2023 = dataset[dataset['Year'] == 2023].groupby('Lead Actor/Actress')['Box Office Revenue (Crores)']
.count().sort_values(ascending=False).head(10)

# x = top_MoviesCollection_2023.index â†’ top_MoviesCollection_2023 (X-axis).
# y = top_MoviesCollection_2023.index â†’ top_MoviesCollection_2023 (Y-axis).
axs[0, 0].bar(top_MoviesCollection_2023.index, top_MoviesCollection_2023)
axs[0, 0].set_title('Top Movies Collection in 2023')
axs[0, 0].tick_params(axis='x', rotation=45)

top_MoviesCollection_2024 = dataset[dataset['Year'] == 2024].groupby('Lead Actor/Actress')['Box Office Revenue (Crores)']
.count().sort_values(ascending=False).head(10)
axs[0, 1].bar(top_MoviesCollection_2024.index, top_MoviesCollection_2024)
axs[0, 1].set_title('Top Movies Collection in 2024')
axs[0, 1].tick_params(axis='x', rotation=45)

axs[1, 1].axis('off')


plt.tight_layout()
```

```python
5 Which Movies by Production Company the most Box Office Revenue (Crores) between 2023 to 2024
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

dataset = pd.read_csv("hindi_movies_data.csv")

dataset_2023_2024 = dataset[(dataset['Year'] >= 2023) & (dataset['Year'] <= 2024)]

(dataset_2023_2024.groupby('Production Company')['Box Office Revenue (Crores)'].sum().sort_values(ascending=False).head(10)
  .plot(kind='bar',title='Top 10 Most Movies Production Company 2023 to 2024'))

plt.tight_layout()
plt.show()
```

8.  what is the Best Director IMDB Rating 2023 to 2024?

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

dataset = pd.read_csv("hindi_movies_data.csv")

dataset_2023_2024 = dataset[(dataset['Year'] >= 2023) & (dataset['Year'] <= 2024)]

(dataset_2023_2024.groupby('Director')['IMDB Rating'].mean().sort_values(ascending=False).head(10)
  .plot(kind='bar',title='Best Director IMDB Rating 2023 to 2024'))

plt.show()
```

9. Top 5 Production Company Box Office Revenue (Crores) percentage?

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

dataset = pd.read_csv("hindi_movies_data.csv")

dataset_2023_2024 = dataset[(dataset['Year'] >= 2023) & (dataset['Year'] <= 2024)]

(dataset_2023_2024.groupby('Production Company')['Box Office Revenue (Crores)'].sum().sort_values(ascending=False).head(5)
  .plot(kind='pie', autopct='%1.1f%%', title='Top 5 Production Company Box Office Revenue percentage'))

plt.show()
```
